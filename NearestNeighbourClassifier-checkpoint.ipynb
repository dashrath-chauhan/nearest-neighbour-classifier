```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#Training sets, containing points and labels(colors)
setX = []
setY = []
colors = ['red','blue','black']
for i in range(20):
    setX.append([round(random.uniform(0,9), 1),round(random.uniform(0,9), 1)])
    setY.append(random.choice(colors))
testPoint = [round(random.uniform(0,9), 1),round(random.uniform(0,9), 1)]
print(setX)
TestSet = np.array(testPoint)
X_Train = np.array(setX)
Y_Train = np.array(setY)
```

```python
#print(X_Train[5,0])
#print(Y_Train[5])
print(X_Train[:,0])
print(X_Train[:,1])
```

```python
plt.figure()
plt.scatter(X_Train[:, 0], X_Train[:, 1], s=170, color = Y_Train[:])
plt.scatter(TestSet[0], TestSet[1], s=170, color = 'green')
plt.title('Before Classification')
plt.show()
```

```python
#function that classifies the nearest neighbour

def dist(x,y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
#for each point in X_Train, we compute the distance to TestSet
num = len(X_Train)
distance = np.zeros(num)
for i in range(num):
    distance[i] = dist(X_Train[i], TestSet)
print(distance)
```

```python
#choose the point in X_train with minimum distance to the new test point
minimum = np.around(np.argmin(distance), decimals=1)
print(Y_Train[minimum])
plt.figure()
plt.scatter(X_Train[:, 0], X_Train[:, 1], s=170, color = Y_Train[:])
plt.scatter(TestSet[0], TestSet[1], s=170, color = Y_Train[minimum])
plt.title('After Classification')
plt.show()
```
